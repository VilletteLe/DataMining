# Khám phá và xử lý dữ liệu Marketing Campaign

## 1. Import thư viện và nạp dữ liệu vào notebook

```python
import pandas as pd
```

```python
marketing_data = pd.read_csv("marketing_campaign.csv", sep=None, engine="python")
marketing_data = marketing_data[['ID','Year_Birth', 'Education',
'Marital_Status','Income','Kidhome', 'Teenhome',
'Dt_Customer', 'Recency','NumStorePurchases',
'NumWebVisitsMonth']]
```

```python
print("5 dòng đầu tiên của dữ liệu: ")
marketing_data.head(5)
```

```python
print("Kiểu dữ liệu mỗi cột: ")
display(marketing_data.dtypes)
```

```python
print("Số dòng, số cột của dữ liệu: ")
display(marketing_data.shape)
```

## 2. Loại bỏ dữ liệu trùng lặp

### Xóa các dòng trùng lặp 

```python
duplicates_count = marketing_data.duplicated().sum()
print("Số dòng trùng lặp: ", duplicates_count)
```

```python
print("Các dòng trùng lặp:")
display(marketing_data[marketing_data.duplicated()])
```

```python
marketing_data_duplicate = marketing_data.drop_duplicates()
```

### Xóa dòng có chỉ số (index) là 1

```python
marketing_data.drop(labels=[1], axis=0)
```

### Xóa cột "Year_Birth"

```python
marketing_data.drop(labels=['Year_Birth'], axis=1)
```

## 3. Thay thế dữ liệu và thay đổi định dạng của dữ liệu

Thay thế các giá trị trong cột "Teenhome" bằng:
 - "has no teen" nếu giá trị bằng 0.
 - "has teen" nếu giá trị bằng 1.
 - "has teen" nếu giá trị bằng 2.

```python
marketing_data['Teenhome_replaced'] = marketing_data['Teenhome'].replace([0,1,2],['has no teen','has teen','has teen'])
marketing_data.head(5)
```

```python
display(marketing_data[['Teenhome', 'Teenhome_replaced']].head())
```

### Xử lý giá trị bị thiếu
  - Điền giá trị bị thiếu của cột "Income" bằng 0

```python
print("Số lượng giá trị thiếu trong mỗi cột:")
display(marketing_data.isnull().sum())
```

```python
marketing_data['Income'] = marketing_data['Income'].fillna(0)
display(marketing_data.isnull().sum())
```

### Đổi kiểu dữ liệu
  - Đổi kiểu dữ liệu của cột "Income" từ float sang int

```python
marketing_data['Income_changed'] = marketing_data['Income'].astype(int)
print(marketing_data['Income_changed'])
```
