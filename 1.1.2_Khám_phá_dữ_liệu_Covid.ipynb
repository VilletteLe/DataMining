# Khám phá dữ liệu COVID

## Tính mean, median, mode, variance, standard deviation, range, percentile, quartile, interquartile range (IQR) sử dụng thư viện numpy và stats trên tập dữ liệu COVID.

### Import thư viện

```python
import numpy as np
import pandas as pd
from scipy import stats
from scipy.stats import skew, kurtosis
```

### Nạp dữ liệu

```python
covid_data = pd.read_csv("owid-covid-data.csv")
covid_data = covid_data[['iso_code','continent',
'location','date','total_cases','new_cases']]
```

### Một số thông tin về dữ liệu

```python
print("5 dòng đầu dữ liệu: ")
display(covid_data.head(5))
```

```python
print("5 dòng cuối dữ liệu: ")
display(covid_data.tail(5))
```

```python
print("Kiểu dữ liệu của mỗi cột: ")
display(covid_data.dtypes)
```

```python
print("Số dòng, số cột của dữ liệu: ")
display(covid_data.shape)
```

 ### Loại bỏ các giá trị NaN có trong dữ liệu để các phép tính toán ra kết quả đúng và có ý nghĩa thống kê

```python
data = covid_data.dropna()
```

### Biến "new_cases"

1. Mean (Trung bình)

```python
data_mean = np.mean(data["new_cases"])
print(data_mean)
```

2. Median (Trung vị)

```python
data_median = np.median(data["new_cases"].dropna())
print(data_median)
```

3. Mode (Giá trị xuất hiện nhiều nhất)

```python
data_mode = stats.mode(data["new_cases"], keepdims=True)[0][0]
print(data_mode)
```

4. Variance (Phương sai)

```python
data_variance = np.var(data["new_cases"])
print(data_variance)
```

5. Standard Deviation (Độ lệch chuẩn)

```python
data_sd = np.std(data["new_cases"])
print(data_sd)
```

6. Min - Max (Giá trị nhỏ nhất - Giá trị lớn nhất)

```python
data_max = np.max(data["new_cases"])
print("Max: ", data_max)
data_min = np.min(data["new_cases"])
print("Min: ", data_min)
```

7. Range (Khoảng giá trị)

```python
data_range = data_max - data_min
print(data_range)
```

8. Percentiles (Phân vị)

```python
data_percentile25 = np.percentile(data["new_cases"],25)
print(data_percentile25)
data_percentile50 = np.percentile(data["new_cases"],50)
print(data_percentile50)
data_percentile60 = np.percentile(data["new_cases"],60)
print(data_percentile60)
data_percentile75 = np.percentile(data["new_cases"],75)
print(data_percentile75)
```

9. Quartiles (Q1, Q2, Q3)

```python
data_quartile = np.quantile(data["new_cases"],0.75)
print(data_quartile)
```

10. Interquartile Range (IQR)

```python
data_IQR = stats.iqr(data["new_cases"])
print(data_IQR)
```

11. Skewness (Độ lệch)

```python
data_skew = skew(data["new_cases"], nan_policy='omit')
print(data_skew)
```

12. Kurtosis (Độ lệch) 

```python
data_kurt = kurtosis(data["new_cases"], nan_policy='omit')
print(data_kurt)
```

skew > 0 → lệch phải (có nhiều giá trị lớn đột biến)

kurtosis > 3 → phân bố nhọn hơn bình thường (nhiều outliers)

```python
zero_ratio = (data["new_cases"] == 0).mean() * 100
print(zero_ratio)
```
