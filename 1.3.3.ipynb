3. Thá»±c hÃ nh vá»›i Sweetviz
BÃ¢y giá», hÃ£y má»Ÿ Jupyter Notebook hoáº·c trÃ¬nh soáº¡n tháº£o code yÃªu thÃ­ch cá»§a báº¡n vÃ  báº¯t Ä‘áº§u.

```python
import pandas as pd
import sweetviz as sv

# Äá»c toÃ n bá»™ file dáº¡ng text
with open("marketing_campaign.csv", "r", encoding="utf-8") as f:
    lines = f.readlines()

# Bá» dáº¥u ngoáº·c kÃ©p vÃ  tÃ¡ch cá»™t báº±ng chuá»—i '\t'
clean_lines = [line.strip().strip('"').split("\\t") for line in lines]

# Láº¥y dÃ²ng Ä‘áº§u tiÃªn lÃ m tÃªn cá»™t
header = clean_lines[0]
rows = clean_lines[1:]

# Táº¡o DataFrame
df = pd.DataFrame(rows, columns=header)

print("âœ… Dá»¯ liá»‡u Ä‘Ã£ tÃ¡ch thÃ nh cÃ´ng!")
print("\nðŸ“‹ CÃ¡c cá»™t:")
print(df.columns.tolist())

print("\nðŸ“Š 5 dÃ²ng Ä‘áº§u tiÃªn:")
print(df.head())

# Náº¿u cÃ³ cá»™t 'Response', dÃ¹ng lÃ m target; náº¿u khÃ´ng, bá» qua target_feat
if 'Response' in df.columns:
    my_report = sv.analyze(df, target_feat='Response')
else:
    my_report = sv.analyze(df)

# Xuáº¥t bÃ¡o cÃ¡o
```

```python
import pandas as pd

# Äá»c file, tÃ¡ch báº±ng chuá»—i '\t' tháº­t (2 kÃ½ tá»± '\' vÃ  't')
df = pd.read_csv("marketing_campaign.csv", sep="\\t", engine="python")

# TÃ¡ch tiáº¿p náº¿u váº«n gá»™p táº¥t cáº£ vÃ o 1 cá»™t
if df.shape[1] == 1:
    # Chia cá»™t Ä‘áº§u tiÃªn thÃ nh nhiá»u cá»™t tháº­t
    df = df.iloc[:, 0].str.split(r"\t", expand=True)
    # GÃ¡n láº¡i tÃªn cá»™t tá»« hÃ ng Ä‘áº§u tiÃªn
    df.columns = df.iloc[0]
    df = df[1:].reset_index(drop=True)

print("âœ… Dá»¯ liá»‡u Ä‘Ã£ tÃ¡ch cá»™t thÃ nh cÃ´ng!")
print(df.head())
print("\nðŸ“‹ TÃªn cá»™t:")
print(df.columns.tolist())

# (TÃ¹y chá»n) LÆ°u láº¡i file sáº¡ch
df.to_csv("marketing_campaign_clean.csv", index=False)
```

TÃ­nh nÄƒng 1: PhÃ¢n tÃ­ch toÃ n bá»™ táº­p dá»¯ liá»‡u (analyze)
ÄÃ¢y lÃ  tÃ­nh nÄƒng cÆ¡ báº£n vÃ  máº¡nh máº½ nháº¥t. NÃ³ sáº½ táº¡o má»™t bÃ¡o cÃ¡o tá»•ng quan cho toÃ n bá»™ dataframe cá»§a báº¡n.

```python
import pandas as pd
import sweetviz as sv

# Äá»c dá»¯ liá»‡u, dÃ¹ng tab lÃ m dáº¥u phÃ¢n cÃ¡ch
try:
    df = pd.read_csv("marketing_campaign_clean.csv", sep="\t")
    print("âœ… Táº£i dá»¯ liá»‡u thÃ nh cÃ´ng!")
except FileNotFoundError:
    print("âŒ KhÃ´ng tÃ¬m tháº¥y file 'marketing_campaign_clean.csv'. Kiá»ƒm tra láº¡i Ä‘Æ°á»ng dáº«n.")
    exit()

# Hiá»ƒn thá»‹ 5 dÃ²ng Ä‘áº§u tiÃªn Ä‘á»ƒ kiá»ƒm tra
print("\nðŸ“Š 5 dÃ²ng dá»¯ liá»‡u Ä‘áº§u tiÃªn:")
print(df.head())

# Kiá»ƒm tra cÃ¡c cá»™t Ä‘á»ƒ cháº¯c cháº¯n Ä‘Ã£ tÃ¡ch Ä‘Ãºng
print("\nðŸ“‹ TÃªn cÃ¡c cá»™t:")
print(df.columns.tolist())

# Táº¡o bÃ¡o cÃ¡o Sweetviz (bá» qua target_feat náº¿u chÆ°a cháº¯c cháº¯n cÃ³ cá»™t Response)
my_report = sv.analyze(df)
my_report.show_html('Sweetviz_Report_Marketing.html')
```

TÃ­nh nÄƒng 2: So sÃ¡nh hai táº­p dá»¯ liá»‡u (compare)
TÃ­nh nÄƒng nÃ y cá»±c ká»³ há»¯u Ã­ch khi báº¡n muá»‘n so sÃ¡nh táº­p train vÃ  test trong Machine Learning. á»ž Ä‘Ã¢y, chÃºng ta sáº½ giáº£ láº­p báº±ng cÃ¡ch chia dataframe thÃ nh 2 pháº§n.

```python
# Giáº£ láº­p chia dá»¯ liá»‡u thÃ nh 2 pháº§n: khÃ¡ch hÃ ng cÃ³ thu nháº­p cao vÃ  tháº¥p
# (Trong thá»±c táº¿, báº¡n sáº½ dÃ¹ng vÃ­ dá»¥ nÃ y cho train/test set)
df_high_income = df[df['Income'] > 68000]
df_low_income = df[df['Income'] <= 68000]

# 1. Táº¡o bÃ¡o cÃ¡o so sÃ¡nh hai nhÃ³m
comparison_report = sv.compare([df_low_income, "Thu nháº­p tháº¥p"], [df_high_income, "Thu nháº­p cao"], target_feat='Response')

# 2. Hiá»ƒn thá»‹ bÃ¡o cÃ¡o so sÃ¡nh
comparison_report.show_html('Comparison_Report.html')
```

```python
# So sÃ¡nh cÃ¡c nhÃ³m khÃ¡ch hÃ ng dá»±a trÃªn trÃ¬nh Ä‘á»™ há»c váº¥n ('Education')
# vÃ  xem nÃ³ áº£nh hÆ°á»Ÿng Ä‘áº¿n biáº¿n má»¥c tiÃªu 'Response' nhÆ° tháº¿ nÃ o.
intra_comparison_report = sv.compare_intra(df, df["Education"] == "Graduation", ["Tá»‘t nghiá»‡p", "ChÆ°a tá»‘t nghiá»‡p"], target_feat='Response')

# Hiá»ƒn thá»‹ bÃ¡o cÃ¡o
intra_comparison_report.show_html('Intra_Comparison_Report.html')
```
